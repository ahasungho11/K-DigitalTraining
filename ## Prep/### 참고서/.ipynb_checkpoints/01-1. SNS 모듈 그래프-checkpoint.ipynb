# < matplotlib의 그래프 스타일 >

```python
# matplotlib에서 사용할 수 있는 스타일시트 목록이 출력
import matplotlib.pyplot as plt
style = plt.style.available
print(type(style), len(style), '\n', style)
```

### 1) 막대 그래프 (28가지)

```python
import pandas as pd
import random

series = pd.Series(random.sample(range(100, 1000), 10), 
                index = list('matplotlib'))

for i in range(len(style)):
    
    plt.style.use('{}'.format(style[i]))
    plt.xlabel(style[i], fontsize=20)
    plt.bar(series.index, series.values)
    plt.show()
```

### 2) 꺾은선 그래프 (4가지)

```python
# tplotlib.pyplot을 이용하여 세 가지 유명한 그래프 스타일을 다루어 본다.
import matplotlib.pyplot as plt
#. ggplot style
plt.style.use("ggplot")
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()

# 2. fivethirtyeight style
plt.style.use("fivethirtyeight")
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()

# 3. dark background
with plt.style.context("dark_background"):
     plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()

# 4. seaborn pastel
with plt.style.context("seaborn-pastel"):
     plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

# < 꺾은선 그래프 각각 색깔 지정하기 >

```python
import numpy as np

x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
plt.style.use("ggplot")
plt.plot(x, np.sin(x), color = 'red', label = 'y = sin x')
plt.plot(x, np.cos(x), color = 'blue', label = 'y = cos x')
plt.plot(x, x, color = 'green', label = 'y = x')

# plot의 'label'에 써둔 것으로 범례 자동 지정
plt.legend(loc = 2)

plt.show()
```

```python
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)

# 위 그래프와 같은 결과 출력
# plt.style.use("fivethirtyeight")
with plt.style.context("dark_background"):
    plt.plot(x, np.sin(x), color = 'red')
    plt.plot(x, np.cos(x), color = 'blue')
    plt.plot(x, x, color = 'green')

# 범례 텍스트를 따로 지정
plt.legend(['y = sin x', 'y = cos x', 'y = x'], loc='best')

plt.show()
```
