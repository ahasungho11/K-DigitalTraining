```python
# import re 불러와야 하지요
import re
# compile() 사용 안함
# match(pattern, 문자열)
m = re.match('[a-z]', 'Python')  # 알파벳 소문자
print(m)
print(re.search('apple', 'I like apple!'))
# 출력하는 형태 보기 => 인덱스로 나오는 것 위주
```

```python
# findall() 메소드
p = re.compile('[a-z]+')  # 소문자 패턴
print(p.findall('Life is too short'))
# 찾아서 리스트에 담아서 반환
# 조건에 맞게 대문자는 빼고 반환
```

```python
# findall은 매칭이되는 자료를 리스트로 담아 주니까 바로 뽑아쓸 수 있겠지용
p = re.compile('[a-z]+')
result = p.search('I like apple 123')
print(result)

result = p.findall('I like apple 123')
print(result)
print(result[0])
```

```python
# match() 메소드
tel_checker = re.compile('^(\d{2,3})-(\d{3,4})-(\d{4})$')

print(tel_checker.match('02-123-4567'))
print(tel_checker.match('053-950-45267'))  # 마지막 자리수가 틀림
print(tel_checker.match('02123-4567'))     # 대시(-)가 없음
```

```python
# groups() -> 매칭된 결과를 튜플로 출력
m = tel_checker.match('02-123-4567')
print(m.groups())                # gorups()는 튜플 형태로 출력하는 것만 다름
print('group(0):', m.group(0))   # groups() = groups(0): 02-123-4567
print('group(1):', m.group(1))   #                      (1) (2)  (3)
print('group(2,3):', m.group(2,3))
print()
print('start():', m.start())     # 매칭된 전체 문자열의 시작 인덱스
print('end(0):', m.end())        # 마지막 인덱스 + 1 (range의 end 파라미터 부분과 같다고 보면 됨)

```

```python
# 전방(위 -> 아래 : 위 // 왼 -> 오른 : 왼)
# 전방 긍정 탐색 : (?=패턴)
# - 패턴과 일치하는 문자열을 만나면 패턴 앞의 문자열 반환

# 전방 긍정 탐색 (.: 임의의 문자, )
lookahead1 = re.search('.+(?=won)', '1000 won')  # 패턴(won) 앞의 문자열을 가지고 옴
print(lookahead1)

lookahead2 = re.search('.+(?=log:)', '2022-07-01 00:00:01 ABC.log: 전방탐색')
print(lookahead2)

# 전방 부정 탐색
lookahead3 = re.search('\d{4}(?!-)', '010-1234-5678')
print(lookahead3)
```

```python
# 후방 긍정 탐색
lookbehind1 = re.search('(?<=log:).+', '2022-07-01 00:00:01 ABC.log: this is python')
print(lookbehind1)

lookbehind2 = re.search('(?<=:).+', 'USD: $51')
print(lookbehind2)

# 후방 부정 탐색(\b: 공백(blank)검색)
lookbehind3 = re.search('\\b(?<!\$)\\d+\\b', 'I paid $30 for 100 apples.')
print(lookbehind3)

# r'문자열'
lookbehind4 = re.search(r'\b(?<!\$)\d+\b', 'I paid $30 for 100 apples.')
print(lookbehind4)
```

```python
#5
# 앞서서는 <a>만 따온 것이지만, 
# 항목 링크(연관 기사 링크)만 가져옴
# 정규식 : '^(/wiki/)((?!:).)*$'
# -> 전방 부정
# 항상 줄여놓고 변수로 받아야 데이터가 줄어들지
```

```python
html = urlopen('https://en.wikipedia.org/wiki/Kevin_Bacon')
bs = BeautifulSoup(html, 'html.parser')
body_content = bs.find('div', {'id':'bodyContent'})

# ^(정규식 시작)... $(정규식 끝)
# (/wiki/): '/wiki/를 포함
# ((?!:).)*: ':' 콜론이 없는 문자열 및 임의의 문자가 0회 이상 반복되는 문자열 검색

pattern = '^(/wiki)((?!:).)*$'

for link in body_content.find_all('a', href=re.compile(pattern)):
    if 'href' in link.attrs:
        print(link.attrs['href'])
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import datetime
import random
import re

random.seed(datetime.datetime.now())
random.seed # 난수 생성

def getLinks(articleUrl):
    html = urlopen('https://en.wikipedia.org{}'.format(articleUrl))
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find('div', {'id':'bodyContent'}).find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))

links = getLinks('/wiki/Kevin_Bacon')
while (len(links)) > 0:
    newArticle = links[random.randint(0, len(links)-1)].attrs['href']
    print(newArticle)
    links = getLinks(newArticle)
```

### 같은 페이지를 두 번 크롤링 하지 않기

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re

pages = set() # 세트 선언
def getLinks(pageUrl):
    global pages
    html = urlopen('https://en.wikipedia.org{}'.format(pageUrl))
    bs = BeautifulSoup(html, 'html.parser')
    for link in bs.find_all('a', href = re.compile('^(/wiki/)')):
        if 'href' in link.attrs:
            if link.attrs['href'] not in pages:
                newPage = link.attrs['href']
                print(newPage)
                pages.add(newPage)
                getLinks(newPage)
getLinks('')
```
