## (부제 : 위기의 곡물자급률, 느슨해진 우리나라 농업에 긴장 심기)

```python
# DF 불러오기
selfDF = pd.read_csv('곡물자급률(1957~2020).csv', encoding = 'euc_kr')

# 데이터 확인
selfDF.info()
```

```python
selfDF.head()
```

```python
# 필요한 행만 출력(2016년~2020년, 최근 5년)
self2DF = selfDF.iloc[59:]
self2DF
```

```python
self2DF['연도'] = [2016,2017,2018,2019,2020]
self2DF
```

```python
self2DF.set_index('연도')
self2DF
```

```python
fig = plt.figure(figsize=(10, 8))
plt.rc('font',family='Malgun Gothic')
plt.plot(self2DF['연도'], self2DF['보리쌀'], marker='o')
plt.plot(self2DF['연도'], self2DF['밀'], marker='s')
plt.plot(self2DF['연도'], self2DF['옥수수'], marker='s')

plt.legend(('보리쌀','밀','옥수수',),fontsize=15)
plt.title('<우리나라 곡물자급률>', size=25)
plt.xticks(range(2016,2021), rotation=70, size=15)

plt.show()
```

- 우리나라 곡물자급률 중 괜찮은 수치를 가진 '보리쌀'과 우크라이나 생산에 의존하는 '밀','옥수수' 비교
- '보리쌀'은 떨어지는 추세이지만, 그래도 자급률에 크게 영향 없음.
- 하지만 '밀','옥수수'는 둘 다 0점 대를 유지하고 있기 때문에 비상!

```python
self2DF
```

```python
# 정규화
d=[]
for col in self2DF.columns:
    c=[]
    for i in self2DF.index:
        c.append(round((self2DF.loc[i][col] - self2DF[col].mean()) / self2DF[col].std(), 2))
    d.append(c)
d
```

```python
d0=pd.Series(d[0])
d1=pd.Series(d[1])
d2=pd.Series(d[2])
d3=pd.Series(d[3])
d4=pd.Series(d[4])
d5=pd.Series(d[5])
d6=pd.Series(d[6])

plt.figure(figsize=(20,10))
plt.rcParams['axes.unicode_minus']=False
# 필요한 밀, 옥수수만 
plt.plot(self2DF['연도'], d2,marker='o')
plt.plot(self2DF['연도'], d3,marker='o')
plt.xticks(range(2016,2021),size=20)
plt.legend(('밀','옥수수',),fontsize=15)
plt.title("<밀, 옥수수 곡물자급률>",size=20)

plt.show()
```

 ### <span style='color:Blue'> 최종 결론 및 해결 방안 : 
(1) 개인적 해결방안 : 팜유 세계 최대 수출국이면서도 식용유 품귀현상을 겪는 인도네시아에서는 '식용유 없이 요리하는 법' 시연회를 개최하는 등 튀기지 말고 끓이거나 쪄서 먹자는 캠페인을 벌이고 있다. 우리나라도 가정에서는 튀거나 볶는 음식 대신 찌거나 삶는 음식을 먹는 캠페인을 유도해야 한다.

(2) 국가적 해결방안 : 우리나라는 세계식량안보 지수 순위가 OECD 국가 중 최하위권에 속하기 때문에 국가에서 법적으로 구속해야한다는 의견. 전쟁 뿐만 아니라 기후 위기로 인해 자급자족에 대한 위기가 더욱 심화되고 있기 때문이다. 우리랑 비슷한 순위에 있는 일본은 국가에서 지급가능한 곡물을 늘리기 위해 논을 밀 경작지로 전환하는 정책을 보였다. 그래도 일본은 미국에 곡물회사를 가지고 있어 우리나라보다 더 나은 상황이라고 볼 수 있다. 우리나라도 국내에 버려지는 농지를 활용하는 방법을 찾거나, 해외 식량공급망 확보에 주의를 기울여야한다.
