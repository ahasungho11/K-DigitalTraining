## 미국 금리변화에 따른 (가계 & 기업) 부채 변화
 - 자료출처 : 한국은행

```python
import pandas as pd
import matplotlib.pyplot as plt
import platform
data=pd.read_csv('가계신용동향(분기별).csv', encoding='euc_kr')
data.head()
```

```python
debt_columns=[]
for i in range(len(data.columns)):
    if i == 0 or i==1:
        debt_columns.append(data.columns[i])
    else:
        debt_columns.append(data.columns[i][:4]+'/' +data.columns[i][4:])
    
```

```python
data.columns=debt_columns
data
```

```python
data.columns[2:]
```

```python
home_credit=data.iloc[0][18:]
home_debt=data.iloc[3][18:]
sell_credit=data.iloc[6][18:]
print(home_credit,home_debt,sell_credit)
```

```python
home_credit=home_credit.astype('int64')
home_debt=home_debt.astype('int64')
sell_credit=sell_credit.astype('int64')
print(home_credit ,home_debt,sell_credit)
```

```python
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='AppleGothic')


plt.figure(figsize=(15,10))
plt.title('가계신용동향 (단위: 조원)', fontsize=15)
plt.ylim(80,1900)

plt.plot(home_credit.index,home_credit.values, 'r-',marker='s',label='가계신용')
plt.plot(home_debt.index,home_debt.values, 'b-',marker='s',label='가계대출')
plt.legend(fontsize=15)
plt.yticks(fontsize=15)
plt.xticks(rotation=50, fontsize=15)
plt.twinx()

plt.plot(sell_credit.index,sell_credit.values, 'g-',marker='s',label='판매신용')
plt.legend(loc='lower right',fontsize=15)

plt.yticks(fontsize=15)
plt.show()
```

## 가계신용 분석
    1. 가계신용= 가계대출 +판매신용
       - 가계대출:  순수가계에 대한 금융기관 대출 포괄
       - 판매신용: 제화(물품)의 판매(생산)자나 서비스제공자가 제공하는 외상(신용)거래를 포괄
