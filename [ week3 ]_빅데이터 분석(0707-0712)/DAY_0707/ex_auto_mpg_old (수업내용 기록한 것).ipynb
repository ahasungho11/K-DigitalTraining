### 3-3-1) 방법 : 컬럼의 최대값의 절대값으로 전체 데이터 각각을 나누기

```python
data = auto_mpg['weight']
mpg_max = auto_mpg['weight'].max()
mpg_min = auto_mpg['weight'].min()
```

```python
auto_mpg['weight_nor'] = auto_mpg['weight'] / abs(auto_mpg['weight'].max())

# 범위로 낸다면~
auto_mpg['weight_nor'].min(), auto_mpg['weight_nor'].max()
```

### 3-3-2) 방법 : (컬럼의 데이터-최소값) / (최대값 - 최소값)

- 컬럼의 최대값 => max(), 최소값 => min()

```python
numerator = auto_mpg['weight'] - mpg_min
denominator = mpg_max - mpg_min

auto_mpg['weight_nor_3'] = numerator / denominator
auto_mpg['weight_nor_3'].min(), auto_mpg['weight_nor_3'].max()
```

```python
auto_mpg[['weight', 'weight_nor_2', 'weight_nor_3']]
```

### 3-3-3) 방법 : (컬럼의 데이터 - 평균) / 표준편차 ==> Z-score

```python
(data - data.mean()) / data.std()
```

### 3-3-4) 방법 : (컬럼의 데이터 - 최소값) / (최대값-최소값)

```python
(data - mpg_min) / (mpg_max - mpg_min)
```
