```python
# 모듈 로딩
import pandas as pd

# 파일 관련 변수들
DIR = '../Data/'
FILE = DIR+'stock-data.csv'

FILE
```

```python
# 1) DSV FILE => DataFrame 로딩
stockDF1 = pd.read_csv(FILE)
stockDF1.info()
```

```python
# 함수로 만들어 놓으면, 이렇게 확인하고, 결측치 체크하는 등의 작업을 하게끔 만들어 놓고 쓰면 되겠지
```

```python
# 3) 전처리 => NaN, Duplicate, DateType
# Date 컬럼이 Object니까, datetime64로 바꿔

# 방법!(1) Series.astype()
# astype에는 inplace파라미터가 없으니까 담아줘
stockDF1['Date'] = stockDF1['Date'].astype('datetime64')
stockDF1.info()
```

```python
# 방법(2) pandas.to_datetime()
stockDF1['Date'] = pd.to_datetime(stockDF1['Date'])
stockDF1.info()
```

```python
stockDF1.head()
```

```python
# 밑에서 인덱스로 하는 바람에 안되는 것 => 다시 해보기
# 반드시 datetime으로 바꿔주고 실행해야 dt를 쓸 수 있다는 것!!

# 특정 데이터 추출
# 7월 데이터만 추출
stockDF1['Date'].dt.month == 7
stockDF1[stockDF1['Date'].dt.month == 7]
```

```python
stockDF1
```

```python
# 6월 20일부터 25일 데이터만 추출
# 내대로, 캡처본 2가지

# i) 내대로

a = stockDF1['Date']

stockDF1[(a.dt.day >= 20) & (a.dt.day <= 25) & (a.dt.month == 6)]
# 단, 'and'와 '|'는 안됨 => '&'가 되는 것이다.
```

```python
# ii) 변수를 줘서
day = stockDF1['Date'].dt.day
month = stockDF1['Date'].dt.month
m_d = (month == 6) & (20 <= day) & (25 >= day)
stockDF1[m_d]
```

```python
# Index 변경 => Date 컬럼을 인덱스 설정 => set_index(컬럼명)
stockDF1 = stockDF1.set_index('Date')
# set_index inplace나 다시 넣는 것 잊지 말기
```

```python
stockDF1.info()
```

```python
stockDF1.index
```

```python
# '문자열'로 뽑은 전과는 달리 '인덱스'로 뽑아보는 것 실습
# 6월 20일부터 25일까지 데이터만 추출 => df.loc[], df.iloc[]

# i) 위의 stcckDF1의 index를 뽑은 것들 중에서 긁어와서 슬라이싱!
stockDF1.loc['2018-06-20':'2018-06-25']
```

```python
# ii) for문을 사용해서
a = []
for d in stockDF1.index:
    if (d.month == 6) & (d.day >= 20) & (d.day <= 25):
        a.append(stockDF1.loc[d])

result = pd.DataFrame(a)
result
# 하나씩 뽑으면 Timestamp가 되
```
