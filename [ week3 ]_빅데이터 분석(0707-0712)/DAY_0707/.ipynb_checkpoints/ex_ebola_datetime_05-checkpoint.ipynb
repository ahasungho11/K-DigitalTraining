```python
# csv 읽어올 때
# 1) 구분자가 콤마(,)가 맞는가
# 2) 어딘가에 적혀있겠찌
# 3) 인덱스로 쓸만한 컬럼이 있는가
```

```python
# 모듈 로딩
import pandas as pd

# 파일 관련 변수들
DIR = '../Data/'
FILE = DIR+'country_timeseries.csv'

FILE
```

```python
# 1) CSV => DataFrame 로딩
ebolaDF1 = pd.read_csv(FILE)
#nrows : 첨부터 원하는 행의 개수를 적으면 됨
ebolaDF1
```

```python
# 2) 데이터 확인
ebolaDF1.info()
```

```python
# 3) 전처리

# 분석의 목적도 생각하면서, 결측치의 어느 부분을 날릴 것인가, 수정할 것인가도 생각해줄 수 있도록
# -> 파라미터 뭐가 있는지도 봐주면서 쇽쇽 떠올릴 수 있으면 좋으니 -> 즉각적 판단

# 3-1) NaN 데이터 처리 => 어떤 데이터인지 확인부터 해야 처리 방식 결정
# 한 국가당 건수가 기재되어 있지 않은 컬럼을 일정 NaN 데이터가 있는 곳을 없애주고자함
# => info를 확인해서 넣어줄  데이터가 어떤 상태인지 확인할 것

ebolaDF1.fillna(method = 'bfill', inplace=True)
ebolaDF1.fillna(0)
```

```python
# 3-2) Data Type 정리
# 금일 날짜를 뽑아서 빼주기 위해 같은 데이터 타입인 datetime으로 변환
ebolaDF1['Date'] = pd.to_datetime(ebolaDF1['Date'])
ebolaDF1.info()
```

```python
a = ebolaDF1.loc[121]['Date']
a
```

```python
# 4) 데이터 가공
# Ebola 바이러스 발병 후 몇일(Day)이 지났는지에 대한 컬럼 추가
# 현재 날짜 뽑아서
from datetime import datetime
currentDT = datetime.today()  # 얘는 클래스 메서드
b = currentDT - a
b
# Timedelta로 나옴
```

```python
c = b.days
c
# timedelta에 해당하는 Attribute를 볼 수 있도록 .을 찍고 tab!!
```
