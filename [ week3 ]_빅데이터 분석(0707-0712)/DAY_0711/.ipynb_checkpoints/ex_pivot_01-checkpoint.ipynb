# 피벗(Pivot) 실슬

- 대량의 데이터(DataFrame)에서 원하는 컬럼을 선택해서 Index, Column, Values로 설정  
- 새로운 DF 객체 생성  
- 함수 및 메서드  
    => pandas.pivot(), pandas.DataFrame.pivot()  
    => pandas.pivot_table(), pandas.DataFrame.pivot_table() 

```python
# 247~
# pivot => 필요한 데이터 컬럼만 뽑아서 부분부분 정해서 원하는 형태로 만듦
# => 행 인덱스, 열 인덱스, 데이터 값, 데이터 집계 함수
# 정제 작업 이후 수행
```

```python
# 모듈 로딩
import pandas as pd
import numpy as np

# 데이터 생성
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
df
```

```python
# 데이터 구성 확인
# 1. 데이터 전처리 => 정제(Cleaning)
# -> 결측치 처리
df.isnull().sum()
```

```python
# -> 중복데이터 제거
df.duplicated()
```

```python
# 2. pivot으로 원하는 DF 생성
myDF = pd.pivot(df, index='foo', columns='bar', values='baz')
myDF
```

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
df
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],
                    columns=['C'], aggfunc=np.sum)
table
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],
                    columns=['C'], aggfunc=np.sum, fill_value=0)
table
```

```python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'],
                    aggfunc={'D': np.mean,
                             'E': np.mean})
table
```
