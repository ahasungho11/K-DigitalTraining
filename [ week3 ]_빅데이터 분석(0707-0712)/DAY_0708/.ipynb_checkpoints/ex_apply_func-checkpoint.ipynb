### apply(함수명, axis=방향/축) 실습

- apply의 axis=0은 열을 뜻함
- 원소/요소에 함수 적용하기
- 입력 : Series의 요소 또는 DataFrame의 Series (즉, 행 또는 열)
 -> axis =0 (행을 나타내는데, 시리즈 전체를 나타내기 위해서는 열의 값이 모두 필요하다고 생각할 것

```python
import numpy as np

# [ 기본 ] axis = 0     =>     row의 컬럼마다 함수 적용 (세로로 죽)
subDF.apply(np.sum)
```

```python
# [ 수정 ] axis = 1     =>     column의 로우마다 함수 적용 (가로로 죽)
subDF.apply(np.sum, axis = 1)
```

```python
# 사용자 정의 함수
def addValue(x):
    print(f'[DEBUG]--------- \n{x}')
    return x+x

def addTwoValue(x, y):
    print(f'[DEBUG]--------- \n{x}, {y}')
    return x+y
```

```python
subDF.apply(addValue)  # axis = 0임 (기본값)
```

```python
subDF.apply(addValue, axis=1)
```

```python
# Series.apply() ---> 원소 단위
subDF['Age'].apply(addValue)

# Series는 axis = 1이 없으니 무의미
```

```python
# subDF에 있는 값을 x인자로 받고, y인자는 줬음
subDF.apply(addTwoValue, y=2000)
```

```python
# 사용자 정의 함수
def missingValueCheck(sr):
    print(f'[DEBUG]--------- \n{sr}')
    return sr.isnull()

def missingValueSum(sr):
    print(f'[DEBUG]--------- \n{sr}')
    return sr.isnull().sum()
```

```python
subDF.apply(missingValueCheck)
```

```python
subDF.apply(missingValueSum, axis=1)
```
