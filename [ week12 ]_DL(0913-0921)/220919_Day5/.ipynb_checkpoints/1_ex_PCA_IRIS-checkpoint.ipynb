```python
class skelarn.decomposition.PCA
n_components=None : 
개수를 줄수도, 비율을 줄 수도
주성분을 뽑아냈을 때, 해당 성분을 만족하는 data가 n개는/n%는 되어야 한다

주성분 분석 메서드
PCA.transform( 2D ) : 특성 -> 주성분으로 축소
PCA.inverse_transform(2D) : 주성분 -> 원본 특성으로 복원
=> 차이를 보면 알 수 있겠즹
```

## PCA(주성분분석)
---
- 전체 특성(Feature)에서 주요 특성 압축/융합하는 비지도 학습
- fit() + transform() 2단계 거쳐 주성분 분석 (like scale처럼)

```python
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt       
```

## [1] 데이터 준비
---

```python
irisDF = load_iris
```

```python
# DataFrame으로 갖고 오기
X, y = load_iris(return_X_y=True, as_frame=True)
type(X), type(y)

# 물론 Bunch로 갖고 와서 data와 targt으로 빼도 되고
```

```python
# seaborn 안써도 판다스로 산점도 뿌릴 수 있음
# scatterplot matrix by pandas scatter_matrix()from pandas.plotting import scatter_matrixscatter_matrix(iris,                alpha=0.5,                figsize=(8, 8),                diagonal='kde')plt.show()

# i) 피처들 전체 산점도
scatter_matrix(X,alpha=0.5,figsize=(8, 8))#,diagonal='kde') # => 그래프로 
plt.show()
```

## [2] 데이터 전처리

```python
irisDF[irisDF.columns[:-1]]
```

```python
# 스케일링
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(irisDF[irisDF.columns[:-1]])
X_scaled = scaler.transform(X)
```

## PCA 주성분 분석
---
- iris의 4개 Feature 중에서 주성분 데이터 추출

```python
pca=PCA()
```

```python
# 지정된 데이터에서 주성분 추출 => 
pca.fit(X_scaled)
```

```python
# pca=PCA()의 주성분 분석

# 주성분 행렬
print(pca.singular_values_)
# 주성분에 투영 후, 분산 정도 설명
print(pca.explained_variance_)
# 주성분에 투영 후, 분산에서 떨어져 있는 비율
print(pca.explained_variance_ratio_)
```

```python
# pca1=PCA(n_components=3)의 주성분 분석
pca1=PCA(n_components=3)
pca1.fit(X_scaled)

# 주성분 행렬
print(pca1.singular_values_)
# 주성분에 투영 후, 분산 정도 설명
print(pca1.explained_variance_)
# 주성분에 투영 후, 분산에서 떨어져 있는 비율
print(pca1.explained_variance_ratio_)
```

## 주성분 갯수 지정
---
- i) explained_variance_ratio_의 누적값에서 원하는 분포정도 지정해서 PC갯수 계산
- ii) n_components = 0.0 ~ 1.0 : 주성분에 투영된 데이터 분포 정도. 지정하면 PC갯수 자동 계산

### 방법[1] : explained_variance_ratio_ 누적값으로 PC 차원수 계산

```python
import numpy as np
```

```python
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95)+1   # 0번부터 인덱스 값을 뽑기 때문에 +1
# -> 95% 이상을 만족시켜줄 수 있는 개수를 뽑아달라

print(f'주성분 갯수(차원) : {d}  =>  {d}개 이상')
```

```python
print(pca.explained_variance_ratio_)
print(cumsum)
```

```python
## 다시 지정된 PC 갯수로 주성분 데이터 압축/융합
pca = PCA(n_components=d)
```

```python
pca.fit(X_scaled)
```

```python
irisPCA = pca.transform(X_scaled)
```

```python
irisPCA.shape
```

```python
irisPCA[:5]
```

```python
## 주성분과 데이터 투영 분산 정도 시각화
plt.figure(figsize=(8,5))
plt.plot(cumsum)
plt.plot(d, 0.95, 'ro')
plt.plot([2.0, 2.0], [0, 0.95], 'b--')
plt.axis([0, 4, 0, 1])
plt.xlabel('Dimention')
plt.ylabel('Explainde Variance')
plt.show()
```

### 방법[2] : n_components=비율값
---
- 상대적으로 [1]보다 편함 ([1]은 수치들을 다 확인하는 과정을 거침)
- cumsum값으로 그래프를 그리겠다면 [1]로 해도 되고

```python
pca3 = PCA(n_components=0.95) # 알아서 0.95가 되도록 개수 맞춰줌
```

```python
pca3.fit(X_scaled)
```

```python
pca3.explained_variance_, pca3.explained_variance_ratio_
```

```python
irisPCA3 = pca3.transform(X_scaled)
irisPCA3.shape
```

```python
irisPCADF = pd.DataFrame(irisPCA3, columns=['pca_1', 'pca_2'])
irisPCADF.head()
```

```python
irisPCADF = pd.concat([irisPCADF, y], axis =1)
irisPCADF.head()
```

### 타겟별 시각화

```python
# 타겟별 시각화
target_0 = irisPCADF[irisPCADF.target == 0]
target_1 = irisPCADF[irisPCADF.target == 1]
target_2 = irisPCADF[irisPCADF.target == 2]
```

```python
plt.figure(figsize=(8,5))
plt.scatter(target_0['pca_1'], target_0['pca_2'], c='orange')
plt.scatter(target_1['pca_1'], target_1['pca_2'], c='blue')
plt.scatter(target_2['pca_1'], target_2['pca_2'], c='red')
plt.legend(['target_0', 'target_1', 'target_2'])
plt.show()
```

### 복원 : inverse_transform

```python
# 변환된 것('irisPCA3 = pca3.transform(X_scaled)')
# => 원래대로 복원
re = pca3.inverse_transform(irisPCA)
re.shape
```
