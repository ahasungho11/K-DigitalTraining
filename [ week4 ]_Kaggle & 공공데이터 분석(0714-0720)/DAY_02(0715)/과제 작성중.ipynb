```python
# 모듈 로딩
import pandas as pd
import platform
import matplotlib.pyplot as plt

# 파일 로딩 -> DF 생성
df = pd.read_csv('subwaytime.csv')
df.head()

# 필요한 컬럼들 뽑기, 컬럼명 변경
commute_df = df.iloc[:, [1, 3, 11, 13]]
commute_df.drop(0,inplace=True)
commute_df = commute_df.rename({'Unnamed: 11':'7-8 하차', 'Unnamed: 13':'8-9 하차'},axis='columns')

# 데이터 타입 바꾸기
commute_df['7-8 하차'] = commute_df['7-8 하차'].astype('int64')
commute_df['8-9 하차'] = commute_df['8-9 하차'].astype('int64')

# array로 만드는 것은 for문 돌아가는 동안에 쓸 것들 list에 담에 두는 것으로 생각


# max_num_list[i] =  max_number
# max_num_list = [0] * 7

# max_station_list[i] = max_station
# max_station_list = [''] * 7

# s_line_list[i] = s_line
# s_line = [''] * 7


max_num_list = []
max_station_list = []
s_line_list = []

for i in range(7):                                                        # 만들어진 배열에 넣기 쉽게끔 0부터 
    line_str = '{}호선'.format(i+1)                                       # 호선 숫자는 맞춰야 하니까 +1
    commute_df_01 = commute_df[commute_df['호선명'] == line_str]
    s_line_list.append(line_str)
    
    row_sum_df = commute_df_01.sum(axis=1)                                
    # df의 해당 행(가로)의 합 (axis=1 이기 때문)
    
    passenger_list = row_sum_df.to_list()

    # 호선 구분 전, 각 행의 합들이 모여 있는 해당 열에서 최대값 찾기
    max_number = row_sum_df.max(axis=0)
    max_num_list.append(max_number)

    max_index = row_sum_df.idxmax()
    
    max_line, max_station = df.iloc[max_index, [1, 3]] # [1]: 호선,[3]: 지하철역명 // 원본 DF인 df에서 수행해야 함
    max_station_list.append(max_station)
    
    print('출근 시간대 {0} 최대 하차역은 {1}역이고 하차인원은 {2:,}명 입니다.'.format(max_line,max_station, max_number))
```

```python
# 한글 폰트 사용
import matplotlib.font_manager as fm
if platform.system() == 'Windows':
    font_name = fm.FontProperties(fname='c:\Windows\Fonts\malgun.ttf').get_name()
    plt.rc('font', family= 'Malgun Gothic', size=8)
    
else:
    plt.rc('font', family='AppleGothic', size=8) # 한글 폰트 사용 For Mac OS

# x축 라벨 준비
label = []
for i in range(7):
    label.append(s_line_list[i]+':'+max_station_list[i])
    
# 막대 그래프 그리기
import matplotlib.pyplot as plt

plt.figure(dpi = 120)                                                      # 그래프 크기 조절
plt.bar(label, max_num_list, color='dodgerblue')                           # x축 데이터, y축 데이터, 막대 색깔
plt.title("출근 시간대 지하철 노선별 최대 하차 인원 및 하차역", size=12)   # 그래프 제목
plt.xticks(rotation = 80, fontsize = 11)
plt.yticks(fontsize = 11)
plt.show()
```
