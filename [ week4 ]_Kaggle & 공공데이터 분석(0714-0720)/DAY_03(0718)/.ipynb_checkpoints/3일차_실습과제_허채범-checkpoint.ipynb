```python
# 모듈 로딩
import csv
import matplotlib.pyplot as plt
import platform

city_list = ['중구','동구','서구','남구','북구', '수성구', '달서구']    # csv 특성상 순서대로 적어줘야
```

```python
# 파일 읽기
f = open('gender.csv', encoding = 'euc_kr')
data = csv.reader(f)
header = next(data)

# 구별 총 남자/여자 수치를 담을 리스트
male_list = []      
female_list = []

for i in range(7):
    for row in data:
        district = '대구광역시 ' + city_list[i]                     # '대구광역시 + OO'이 포함된 자료만 출력
        if district in row[0]:                
            male_list.append(int(row[104].replace(',', '')))       # 해당 구별 총 남자수를 담아
            female_list.append(int(row[207].replace(',', '')))     # 해당 구별 총 여자수를 담아
            break                                                  # 해당하는 '구'의 전체값만 뽑기 위해 한 행만 수행하고 중지7
                                                                   # 도시 하위 목록이 많음. 처음에 나오는 데이터가 전체 총합
            
f.seek(0)   # 파일 제일 처음부터 시작하지 않으면 '달서구 성당동'에 해당하는 총 남자/여자수를 담게 되므로 

for row in data: 
    district = '대구광역시 '                                        # '대구광역시 ' 자료만 출력해서 리스트 젤 마지막에
    if district in row[0]:
        male_list.append(int(row[104].replace(',', '')))
        female_list.append(int(row[207].replace(',', '')))
        break
    
f.close()
```

```python
# 그래프 그리는 함수 만들기
def draw_piechart(male_list, female_list):
    
    if platform.system() == 'Windows':
        plt.rc('font', family='Malgun Gothic')
    else:
        plt.rc('font', family='AppleGothic')
        
                                                               # (행, 열, 크기) 순으로 입력하는 것
    fig, axes = plt.subplots(2,4, figsize=(20,14))             # 그래프 하나하나 전체를 의미. for문 밖으로 빼서 한번만 적어줘야 함
    fig.suptitle('대구광역시 구별 남녀 인구 비율', fontsize=25)  # 받은 fig에 대한 것임
                                                               # axes에 대한 것이 없더라도 같이 적어줘야 함
    fig.subplots_adjust(hspace=0.1, wspace=0.3)                # subplot간의 간격 조절
    
    for i in range(8):                                         # 위에서 담은 수치들 8개 모두를 꺼내쓰기 위해서
        population = [male_list[i], female_list[i]]
        color = ['tomato', 'royalblue']
        
        plt.subplot(2, 4, i+1)                                 # plt.subplot(행, 열, 위치)
        plt.pie(population, labels = ['남성','여성'], autopct='%.1f%%',       # 전부 2개 들어가는 것에 맞춰서
                colors=color, startangle=90)
        plt.title('대구광역시 ' + city_list[i], fontsize=20)
        plt.rc('font', size=15)     # 파이차트 안에서의 글자 크기

    plt.show()   # plt.show()는 1번만 호출하면 되니까 위치 중요

    
city_list = ['중구','동구','서구','남구','북구', '수성구', '달서구', '전체'] 
# male_list와 female_list에 담을 때는 '전체'를 마지막에 넣기 위해, 
# city_list를 '달서구'까지만 썼었는데, 이번에는 그래프를 그리기 위해 '전체'를 추가하여 다시 city_list 선언

draw_piechart(male_list, female_list)
# 함수 호출, 받는 인수들 표시 위해 파라미터 적어주기
```

```python
# 배열로 빈리스트 만들어서 넣는 것 
# 대구 구별 남녀 인구 데이터 찾는 함수
def calc_deagu(DEAGU_LIST):
    deagu_male_value_list=[0]*len(DEAGU_LIST) # 각 구별에 맞는 남성 인구 수를 넣기 위한 빈리스트
    deagu_female_value_list=[0]*len(DEAGU_LIST) # 각 구별에 맞는 여성 인구 수를 넣기 위한 빈리스트
    
    for idx in range(len(DEAGU_LIST)): # 중구는 0인덱스, 동구는 1인덱스,...,전체는 7인덱스
        with open('../Data/gender.csv',encoding='euc_kr') as f: # 구의 인덱스를 부를때 마다 파일 열기
            data=csv.reader(f)
            for row in data: # 한줄씩 읽어
                if '대구광역시 '+DEAGU_LIST[idx] in row[0]: # 대구광역시 + 구(동구,중구)가 행정구역에 포함된다면
                    deagu_male_value_list[idx]+=int(row[104].replace(',','')) # 콤마삭제 정수변환 후 맞는 남성 구 인덱스에 덧셈
                    deagu_female_value_list[idx]+=int(row[207].replace(',','')) # 콤마삭제 정수변환 후 맞는 여성 구 인덱스에 덧셈
                    break # 제일 위에있는 한줄만 읽으면 되므로 break
    
    # 리스트의 제일 마지막은 '전체'를 나타내기 위해 합산해서 넣어주는 과정
    deagu_male_value_list[-1]=sum(deagu_male_value_list) # 대구 남성 전체 인덱스는 각 구별의 전체 합과 같음
    deagu_female_value_list[-1]=sum(deagu_female_value_list) # 대구 여성 전체 인덱스는 각 구별의 전체 합과 같음
    return deagu_male_value_list, deagu_female_value_list # 남녀 인구 리스트 반환
```
