```python
import csv
import matplotlib.pyplot as plt
import platform

DIR='../Data/'
f = open(DIR+'gender.csv', encoding='euc_kr')
data = csv.reader(f)

male_count = []
female_count = []

name = ['대구광역시 중구', '대구광역시 동구', '대구광역시 서구', '대구광역시 남구',
    '대구광역시 북구', '대구광역시 수성구', '대구광역시 달서구','대구광역시 전체']
for city in name:
    for row in data:
        if city in row[0]:
            a=int(row[104].replace(',', ''))  # 자리수 문자열 제거 및 숫자로 변환
            b=int(row[207].replace(',', ''))
            male_count.append(a)
            female_count.append(b)
            break    
```

```python
male_count.append(sum(male_count))
female_count.append(sum(female_count))
```

```python
print(male_count)
print(female_count)
```

```python
fig, axes = plt.subplots(2, 4, figsize=(8, 6), sharex=True, sharey=True)
# 전체 그래프의 타이틀 설정
fig.suptitle(" 지역의 남녀 성별 비율", fontsize=15)

if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
else:
    plt.rc('font', family='AppleGothic')

for i in range(8):
    plt.subplot(2, 4, i+1)
    plt.pie([male_count[i],female_count[i]], labels=['남', '여'], autopct='%.1f%%', startangle=90)
    plt.title(name[i])
plt.show()
```
