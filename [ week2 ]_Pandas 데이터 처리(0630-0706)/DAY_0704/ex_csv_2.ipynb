- CSV 파일 읽고, 컬럼별 최고값과 값의 개수 출력

```python
# 모듈 로딩
import pandas as pd

# 파일 관련
DIR_PATH = '../Data/'
FILE_NAME = DIR_PATH+'iris2.csv'
```

```python
# 파일 불러오기 (기본값은 열었을 때, 첫 줄이 컬럼명으로 들어간다)  # 149
irisDF = pd.read_csv(FILE_NAME, header = None)

# 디폴트 값이 0임 -> None로 바꿔줘야
# 데이터인 첫번째 행을 컬럼명으로 안 들어가

# 주피터 상에서 열어서 구분자 확인
# ssv => 스페이스바로 구분
# tsv => tab으로 구분
# 구분자가 csv인데도 ','가 아닌 경우, sep=':' 이런식으로 바꿔줘야 함
```

```python
irisDF
```

```python
# 컬럼명 변경 (로딩할 때, 파라미터를 주었기 때문에)
irisDF.columns = ['일','이','삼','사','오']
```

```python
irisDF
```

```python
# 컬럼별 최고값, 값의 개수
irisDF_des = irisDF.describe()
irisDF_des
irisDF_des.loc['max']
```

```python
# 컬럼별 개수 ('일' 컬럼이 147개인 이유는, 파일을 수정하면서 3개를 지웠기 때문에)
irisDF.value_counts()
```

```python
# value_counts(,dropna=True), 기본값은 True
# False를 하면? -> NaN 값도 나옴 (NaN의 개수를 알고 싶을 때)

irisDF['일'].value_counts(dropna=False)  # NaN의 개수를 알고 싶을 때
```

```python
# 컬럼 값이 NaN, Na인 것도 카운팅 하고 싶을 때, dropna 파라미터 값을 False로
irisDF['일'].value_counts(dropna=True)  # 기본값임
```
