```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Logistic

from sklearn.linear_model import LogisticRegression
```

```python
data, target = load_digits(return_X_y=True)
```

```python
data_train, data_test, target_train, target_test = \
    train_test_split(data, target, random_state=42)
```

```python
data_train.shape, data_test.shape
```

```python
model_lr = LogisticRegression(max_iter=2500)
model_lr.fit(data_train, target_train)
```

```python
# 학습 후 설정된 속성
print('분류 종류 : ', model_lr.classes_)
print('가중치 값 : ', model_lr.coef_)
print('절편 값 : ', model_lr.intercept_)
print('특성 갯수 : ', model_lr.n_features_in_)
```

```python
print('train score :', model_lr.score(data_train, target_train))
print('test score :', model_lr.score(data_test, target_test))
```

```python
ret = model_lr.decision_function([data_test[0]])
print(np.round(ret,3))
```

```python
from scipy.special import softmax, expit
sigValue2 = softmax(ret)
print(np.round(sigValue2,3))
```

```python
# SVC
from sklearn.svm import SVC
```

```python
model_svc = SVC(gamma=0.001)
```

```python
model_svc.fit(data_train, target_train)
```

```python
# 학습 후 설정된 속성
print('분류 종류 : ', model_svc.classes_)
print('절편 값 : ', model_svc.intercept_)
print('특성 갯수 : ', model_svc.n_features_in_)
```

```python
print('train score :', model_svc.score(data_train, target_train))
print('test score :', model_svc.score(data_test, target_test))
```

```python
# 결정함수 y = wx + b
ret = model_svc.decision_function([data_test[0]])
print(np.round(ret,3))
```

```python
sigValue2 = softmax(ret)
print(np.round(sigValue2,3))
```
